```python
import pandas as pd
import numpy as np
```

```python
file1 = pd.read_csv("Data/file1.csv")
```

```python
file2 = pd.read_csv("Data/file2.csv")
```

```python
file3 = pd.read_csv("Data/file3.csv")
```

```python
df =pd.concat([file1 ,file2, file3], axis = 0)
df
```

```python
#merging columns, replace Nan

df.ST.fillna(df.State, inplace=True)
del df['State']

df.GENDER.fillna(df.Gender, inplace=True)
del df['Gender']
df
```

```python
#Dropping 'customer'
df.drop(columns= ['Customer'], inplace=True)
```

```python
#Standardizing Headers
df =df.rename(columns= {'ST':'State', 'GENDER':'Gender'})
```

```python
#check data type
df.info()
```

```python
df['Customer Lifetime Value'] =  pd.to_numeric(df['Customer Lifetime Value'], errors='coerce').fillna(0).astype(int)
df['Number of Open Complaints']= pd.to_numeric(df['Number of Open Complaints'], errors='coerce').fillna(0).astype(int)
```

```python
df.info()
```

```python
#clean up state column
df['State'] = df['State'].replace(['az','wa', 'cali'],['arizona','washington','california'])
df
```

```python
df['State'].value_counts()
```

```python
#Write a function to replace column "State" to different zones. 
#California as West Region, Oregon as North West, Washington as East, and Arizona and Nevada as Central

groups = { "West": ["California"],
    "Northwest": ["Oregon"],
    "East": ["Washington"],
    "Central":["Nevada","Arizona"]
    }
    
df["Zones"] =(df["State"].apply(lambda x: [k for k in groups.keys() if x in groups[k]]).str[0].fillna("Other"))
```

```python
df.drop(columns= ['State'], inplace=True)
```

```python
#removing duplicates
df= df.drop_duplicates()
```

```python
def clean_gender(x):
    if x in ['M', 'MALE']:
        return 'Male'
    elif x in ('F', 'femal'):
        return 'Female'
    else:
        return 'U'
```

```python
#cleaning gender format
df['Gender'] = list(map(clean_gender, df['Gender'])) 
```

```python
#Check and deal with NaN values. 
#Hint:Replacing null values â€“ Replace missing values with mean of the column (for numerical columns)
mean_income = np.mean(df['Income'])
df['Income']= df['Income'].fillna(mean_income)
df
```

```python
#Standardizing the data â€“ Use string functions to standardize the text data (lower case)
df = df.astype(str).apply(lambda x: x.str.lower())
df
```
